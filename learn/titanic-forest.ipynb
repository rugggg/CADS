```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split, KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

```

```python
# load csv
train_csv = pd.read_csv("train.csv")
train_csv.head()
```

```python
test_csv = pd.read_csv("test.csv")
test_csv.head()
```

```python
# select features
df = train_csv[['Pclass','Sex', 'Fare', 'Embarked','Survived']]
df_test = test_csv[['Pclass','Sex', 'Fare', 'Embarked']]

# map binary sex feature to discrete int
mapping = {'male': 0, 'female': 1}
df['Sex'] = df['Sex'].replace(mapping)
df_test['Sex'] = df_test['Sex'].replace(mapping)

mapping = {'S': 0, 'C': 1, 'Q':2}
df['Embarked'] = df['Embarked'].replace(mapping)
df_test['Embarked'] = df_test['Embarked'].replace(mapping)
# drop any NA values
df = df.dropna()

df_test.set_value(152, 'Fare', df_test['Fare'].mean())

mean_ticket = df['Fare'].mean()
# drop survived out of train set
X = df.drop('Survived', axis=1)
# pick out truth values
Y = df['Survived']


#train/test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=1)
```

```python
# init and fit the sklean random forest
clf = RandomForestClassifier(max_depth=5, random_state=0)
clf.fit(X,Y)
```

```python
# cross val
cv_scores = cross_val_score(clf, X, Y, cv=10)

np.mean(cv_scores)
```

```python
# now let's look at the test data and build a submission
df_test.to_csv("df_test")
predictions = clf.predict(df_test)

predictions
```

```python
submission=pd.DataFrame(predictions)
submission.shape
submission.insert(0,'PassengerId',range(892,1310))
submission.columns = ['PassengerId', 'Survived']
submission
submission.to_csv("titanic_submission.csv", index=False)
```
